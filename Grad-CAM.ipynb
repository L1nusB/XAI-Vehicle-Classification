```python
def generateRandomBox(width,height,imgWidth=244,imgHeight=244):
    base = np.zeros((imgHeight,imgWidth))
    box = np.array([1]*width*height).reshape(height, width)

    x, y = np.random.randint(0,[imgWidth-width,imgHeight-height])
    base[y:y+height, x:x+width] = box
    return base
```
